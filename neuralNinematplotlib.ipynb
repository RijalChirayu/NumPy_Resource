```python
import matplotlib.pyplot as plt
import numpy as np
# import pandas as pd
from matplotlib import style
style.use('ggplot')
```

```python
x_data = np.random.random(1000)  * 100  
y_data = np.random.random(1000) * 100
plt.scatter(x_data , y_data , color='Blue' ,  marker = '*' , s = 100 , alpha = 0.3)
plt.show()
```

```python
import random
years = [2006 + x for x in range(16)]
print(years)

weights = np.random.random_integers(60 , 90 , size = (16))
print(weights)

plt.plot(years , weights , c='b' , lw = 3 , linestyle = '--')
plt.show()
```

```python
x = ['C++' , 'Python' , 'Javascript' , 'Go' , 'Java']
y = [16 , 40 , 53 , 21 , 26]
plt.bar(x , y  , color=['red' , 'green' , 'blue' , 'yellow' , 'purple'] , align = 'center' , width= 0.5)
plt.show()
```

```python
normalDistribution = np.random.normal(20 , 1.5 , size=(1000))
plt.hist(normalDistribution  , cumulative= True)
plt.show()
```

```python
langs = ['Python' , 'C++' , 'Java' , 'C' , 'Go' , 'C#']
votes = [50 , 24, 14,6,17 , 10]
explodes = [0 ,0,0,0.2,0,0]
plt.pie(votes , labels=langs , explode= explodes ,autopct='%1.2f%%' , startangle=90)
plt.show()
```

Artificially Created Section for the box plot

```python
first = np.linspace(0 , 10 , 25)
second = np.linspace(10 , 200 , 25)
third = np.linspace(200 , 210 , 25)
fourth = np.linspace(210 , 230 , 25)

data = np.concatenate((first  , second , third, fourth))

plt.boxplot(data)
plt.show()
```

```python
years  = [2014 ,2015 , 2016 , 2017,
          2018 , 2019 , 2020 , 2021]
income = [55 , 62 ,67 ,71,69,78,89,90]

plt.plot(years , income , marker = '.')
plt.title('Years V/s Income')
plt.ylabel('Yearly Income in thousands')
plt.xlabel('Years')
# plt.yticks( [f'${x}k USD' for x in income])
plt.legend('Income')
plt.show()
```

```python
x1 , y1 = np.random.random(100) ,  np.random.random(100)
x2 ,y2 = np.arange(100) , np.random.random(100)
plt.figure(1)
plt.scatter(x1 , y1)

plt.figure(2)
plt.plot(x2 , y2)

plt.show()
```

Subplots

```python
x = np.arange(100)

fig , axs = plt.subplots(2 , 2)
axs[0 , 0].plot(x , np.sin(x))
axs[0,0].set_title('Sine Wave')

axs[0 , 1].plot(x , np.cos(x))
axs[0,1].set_title('Cosine Wave')

axs[1 , 0].plot(x , np.tan(x))
axs[1,0].set_title('Tan Wave')

axs[1 , 1].plot(x , np.log(x))
axs[1,1].set_title('Log Wave')

plt.show()
```
